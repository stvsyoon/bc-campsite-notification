```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from datetime import date
```

```python
# Connect to chrome
ser = Service("C:\Code\chromedriver.exe")
opt = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=ser,options=opt)

# Contains list of available dates
availableDates = []

# loop every weekend in month (--is there a way to not hardcode the index values in a string?)
months = {'Jun':'1221263136414651', 'Jul':'2631364146515661', 'Aug':'2431364146515661'}
for m in months:
  y = 0
  for i in range(4):
    try:
        # Open site
        driver.get("https://camping.bcparks.ca/")  
        # find and search golden ears
        campsite = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.ID, "park-autocomplete"))
        )

        # enter campsite
        campsite.click()
        campsite.send_keys("Golden Ears")
        # campsite.send_keys("Alice Lake")

        # enter dates
        dates = driver.find_element(By.ID, "mat-date-range-input-0")
        dates.click()
        monthSelector = driver.find_element(By.ID,"monthDropdownPicker")
        monthSelector.click()
        month = driver.find_element(By.CSS_SELECTOR, f"[aria-label='{m} 2023']")
        month.click()

        arrivalDate = driver.find_element(By.XPATH, f"//*[@id='mat-datepicker-0']/div/mat-month-view/table/tbody/tr[{months[m][y]}]/td[{months[m][y+1]}]/button")
        arrivalDate.click()
        departureDate = driver.find_element(By.XPATH, f"//*[@id='mat-datepicker-0']/div/mat-month-view/table/tbody/tr[{months[m][y+2]}]/td[{months[m][y+3]}]/button")
        departureDate.click()

        # enter equipment size
        equipment = driver.find_element(By.XPATH,"//*[@id='equipment-field']/div/div[2]/div")
        equipment.click()
        tents = driver.find_element(By.ID,"mat-option-0")
        tents.click()

        # search for campsites
        search = driver.find_element(By.ID,"actionSearch")
        search.click()

        campsiteList = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.ID, "list-view-button-button"))
        )
        campsiteList.click()

        y+=4

        try:
            WebDriverWait(driver, 3).until(
                EC.presence_of_element_located((By.ID, "availability-0"))
            )
            print("found")
            availableDates.append(m + months[m][y:y+4])

        
        except:
            print("not found")

        
    finally:
        print(availableDates)

driver.quit()


```
